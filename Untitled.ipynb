#ML Model Building
#We Choose TV as Independent Variable and sales as a target variable.


```python
x=data[['TV']]
y=data['Sales']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x.shape,x_test.shape,x_train.shape)
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(alpha=1.0),
    "Lasso Regression": Lasso(alpha=0.1),
    "Decision Tree Regression": DecisionTreeRegressor(random_state=42),
    "Random Forest Regression": RandomForestRegressor(n_estimators=100,random_state=42),
    "Gradient Boosting Regression": GradientBoostingRegressor(n_estimators=100,random_state=42)
}

# Fit and predict for each model
for name, model in models.items():
    model.fit(x_train, y_train)  # Fit on training data
    y_pred = model.predict(x_test)  # Predict on test data
    
    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    print(f"{name} Mean Squared Error: {mse:.2f}")
    r2=r2_score(y_test,y_pred)
    print(f"{name} r2 score: {r2:.2f}")
```

```python
#DF WITH ACTUAL VALUES AND PREDICTIONS AND RESULTS
comparision_df=pd.DataFrame({
    'TV': x_test.values.flatten(),
    'Actual': y_test.values,
    'Predicted': y_pred,
    'Residual':y_test.values - y_pred
})
print(comparision_df)
```
